```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_1970"
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text)
table = soup.find('table')
```

```python
table
```

```python
diss = {}
for i in table.findAll('tr')[1:]:
  all = i.findAll('a')
  diss[all[0].string] = [j.string for j in all[1:]]
    
```

```python
for key,value in diss.items():
  print (key,": ",value)
```
